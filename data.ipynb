```python
#packages
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#loading data and taking sum of all sales

data = pd.read_csv('PharmaDrugSales.csv', usecols=['Month', 'Date', 'AceticAcidDerivatives', 'PropionicAcidDerivatives', 'SalicylicAcidDerivatives','PyrazolonesAndAnilides','AnxiolyticDrugs','HypnoticsSndSedativesDrugs','ObstructiveAirwayDrugs','Antihistamines'])
october_data = data[data['Month'] == 10]
october_data['total_sales'] = october_data.iloc[:, 1:].sum(axis=1)


print(data)
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.plot(october_data['Date'], october_data['total_sales'], marker='o', color='orange')
plt.xlabel('Day')
plt.ylabel('Total Sales')
plt.title('Total Sales in October')
plt.grid(True)
plt.show()



```
